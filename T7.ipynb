```python
#Advance python
import numpy as np
```

```python
"""
insert ->np.insert(array,index, value, axis=None)
axis-(none-> means inserting in flattern array),(0->rowwise insert),(1_>column wise inssert)
"""
a1=np.array([[1,2,3,4,5,6,7,8],[2,3,4,5,6,7,8,9]])
print(a1)
a2=np.insert(a1,5,10)    
print(a2)
a3=np.insert(a1,1,[2,3],axis=1)
print(a3)
```

```python
#append
a4=np.array([2,3,4])
print(a4)
a5=np.append(a4,1)
print(a5)
a6=np.append(a3,[2,7])
print(a6)
```

```python
#if we want to join two arrays
b1=np.array([1,2])
b2=np.array([3,4])
print(np.append(b1,b2))
```

```python
#concatenate function
c1=np.array([[1,2,3],[2,3,4]])
c2=np.array([[1,7,3],[9,3,0]])
c3=np.concatenate((c1,c2),axis=0)
c4=np.concatenate((c1,c2),axis=1)
print(c3)
print(c4)
```

```python
"""
delete-> np.delete(array, index, axix=None)
"""
d1=np.array([1,2,3,4,5])
d2=np.delete(d1,1)
print(d2)
```

```python
d3=np.array([[1,2,3],[4,5,6]])
print(d3)
d4=np.delete(d3,0,axis=0)
print(d4)
d4=np.delete(d3,0,axis=1)
print(d4)
```

```python
"""
stacking:
->vstack()   rowise or vertically
->hstack()   columnwise or horizontally
"""
e1=np.array([1,2,3])
e2=np.array([8,9,10])
print(np.vstack((e1,e2)))
print(np.hstack((e1,e2)))
```

```python
"""spliting::
np.split(array,part)
hsplit()
vsplit()
"""
f1=np.array([1,2,3,4,5,6])
print(np.split(f1,2))
```

```python
f2=np.array([[2,3],[4,5]])
print(f2)
print(np.vsplit(f2,2))
```
