```python
import numpy as np
```

```python
temp=[322,667,43,777,44]
total=0
for i in temp:
    total+=i
print(f"average is:{total/(len(temp))}")
```

```python
#list can do the calculation of 5 data easily but in case of millions of data the loop need more time and calculation must be too slow
#numpy solve this problem numpy stand for numerical python
```

```python
t=np.array([322,667,43,777,44])
print(f"Average is {np.mean(t)}")  #this calculation is more faster and simple than above one we dont need looops and by this we can handle millions of data
```
