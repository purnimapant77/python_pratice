```python
import matplotlib.pyplot as plt
product=['a','b','c','d']
sale=[200,700,600,340]
plt.barh(product,sale,color='pink',label='sale 2025')
plt.xlabel('hehe')
plt.title("bar chart")
plt.show()
```

```python
#pichart
import matplotlib.pyplot as plt
values=[100,200,140,234,500]
ll=['dhangadhi','baitadi','kathmandu','nmr','butwal']
col=['green','blue','gray','black','pink']
plt.pie(values, labels=ll, colors=col)
plt.title("cities")
plt.show()
```

```python
#histograms
#used to show the distribution of continous data by grouping
import matplotlib.pyplot as plt
scores=[45,66,76,82,90,85,70,73,68,77,99,63]
plt.hist(scores,bins=5,color='purple',edgecolor='pink')
plt.xlabel("score range")
plt.ylabel("no of students")
plt.title("source distribution of students")
plt.show()
```
