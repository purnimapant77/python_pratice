```python
import matplotlib.pyplot as plt
```

```python
x=['monday','tudeday','wednesday','thrusday','friday']
y=[10,15,7,20,12]
plt.plot(x,y, color="GREEN")
plt.title("Bakery sale in week")
plt.xlabel("day od week")
plt.ylabel("no of cake sold")
plt.show()
```

```python
months=['jan','fev','march','april','may','june','july','agu','sep','oct','nov','dec']
sale=[1000,2000,689,1200,2500,1200,245,784,990,6543,997,1435]
plt.plot(months,sale,color="orange", linestyle='--',marker='o',label="2025 sale for data")
plt.xlabel("months of the year")
plt.ylabel("sale in months")
plt.title("monthly sales")
plt.legend()
plt.grid(color='pink')
plt.ylim(0,5000)
plt.show()
```
